```python
import pandas as pd
import numpy as np
```

```python
roc = pd.read_csv('AUCROC.csv')
roc
```

```python
rocc = pd.read_csv('AUCROC_clustered.csv')
```

```python
roc_uplift = rocc.add(-1*roc, fill_value=0)
roc_uplift
```

```python
import seaborn as sns
cm = sns.light_palette("green", as_cmap=True)

roc_uplift.set_index('Unnamed: 0').style.applymap(lambda x:
    "background-color: red" if x<-0.01 else ("background-color: green" if x>0.01 else "background-color: white"))
```

```python
pr = pd.read_csv('AUCPR.csv')
```

```python
prc = pd.read_csv('AUCPR_clustered.csv')
```

```python
def highlight_max_09(s, props=''):
    return np.where(10 * s >= 9 * np.nanmax(s.values), props, '')
def highlight_max(s, props=''):
    return np.where(s == np.nanmax(s.values), props, '')
pr.set_index('Unnamed: 0').style.apply(highlight_max_09, props='background-color:#00d000', axis=1).apply(highlight_max, props='font-weight:bold', axis=1)
```

```python
roc.set_index('Unnamed: 0').style.apply(highlight_max_09, props='background-color:#00d000', axis=1).apply(highlight_max, props='font-weight:bold', axis=1)
```

```python
pr_uplift = prc.add(-1*pr, fill_value=0)
pr_uplift
```

```python
pr_uplift.set_index('Unnamed: 0').style.applymap(lambda x:
    "background-color: red" if x<-0.01 else ("background-color: green" if x>0.01 else "background-color: white"))
```

```python
roc_uplift.set_index('Unnamed: 0').style.applymap(lambda x:
    "background-color: red" if x<-0.01 else ("background-color: green" if x>0.01 else "background-color: white"))
```

```python
pr_uplift.style.background_gradient(axis=None, cmap='RdYlGn')
```

```python
roc_uplift.style.background_gradient(axis=None, cmap='RdYlGn')
```
