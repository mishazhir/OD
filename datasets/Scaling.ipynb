```python
import numpy as np
from sklearn.preprocessing import StandardScaler
import pandas as pd
```

```python
data = np.load('Classical/02_annthyroid.npz', allow_pickle=True)
X, y = data['X'], data['y']
```

```python
pd.DataFrame(X)
```

```python
s = StandardScaler()
X1 = s.fit_transform(X)
pd.DataFrame(X1)
```

```python
np.savez('Scaled/02_annthyroid.npz', X = X1, y = y)
```

```python
data = np.load('Scaled/02_annthyroid.npz', allow_pickle=True)
X, y = data['X'], data['y']
pd.DataFrame(X)
```
